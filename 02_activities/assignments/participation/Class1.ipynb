```python
# Import necessary libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy

```

```python
np.random.seed(613) # Set a random seed for reproducibility

# Create x values (0 to 49) and random y values (0 to 99)
x=np.arange(50)
y=np.random.randint(0,100,50)

# Create a scatter plot
fig, ax= plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python
# Create a bar plot
fig, ax= plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
# Create a line plot
fig, ax= plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
# Generate new random y values and create a histogram
y=np.random.randint(0,100,50)
fig, ax= plt.subplots(figsize=(5,3))
ax.hist(y)
```

```python
# Create a line plot with labels and title
x=np.arange(50)
y=np.random.randint(0,100,50)
fig, ax= plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('total growth over time')
ax.set_ylabel('total Growth')
ax.set_xlabel('Years since start')
```

```python
# Customize fonts for titles and labels using dictionaries
font1={'family': 'sans-serif', 'color': 'blue', 'size':20}
font2={'family': 'monospace', 'color': 'green', 'size':14}
fig, ax= plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('total growth over time', fontdict=font1)
ax.set_ylabel('total Growth',fontdict=font2)
ax.set_xlabel('Years since start',fontdict=font2)
```

```python
# Scatter plot with customized marker styles
fig, ax= plt.subplots(figsize=(5,3))
ax.scatter(x,y,
           marker='*',
           color='indigo')
```

```python
# Line plot with customized line and marker styles
fig, ax= plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker='*',
           color='indigo',
           linestyle='--',
           linewidth=2)
```

```python
# Line plot with additional marker customizations
x=np.arange(50)
y=np.random.randint(0,100,50)
fig, ax= plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker='*',
           color='indigo',
           linestyle='--',
           linewidth=2,
           markeredgecolor='#FA9359',
           markerfacecolor='#000000')
ax.grid(axis='y', color='red')
```

```python
# Creat a bubble plot  with variable bubble sizes

df = pd.DataFrame({
      'x': np.random.rand(40),
      'y': np.random.rand(40),
      'z': np.random.rand(40),
   })

plt.scatter(df['x'], df['y'], s=df['z']*1000, c='blue',alpha=0.5)
plt.show()
```
