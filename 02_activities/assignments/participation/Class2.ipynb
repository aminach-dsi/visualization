```python
# Import necessary libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
np.random.seed(613) # Set a random seed for reproducibility

# Create x values (0 to 49) and random y1 values (0 to 99) and random y2 values (0 to 99)
x=np.arange(50)
y1=np.random.randint(0,100,50)
y2=np.random.randint(0,100,50)
```

```python
fig, ax =plt.subplots(figsize=(5,3))
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
# Add legend

fig, ax =plt.subplots(figsize=(5,3))
ax.plot(x,y1, label="Person 1")
ax.plot(x,y2, label="Person 2")
ax.legend(loc="lower right")
```

```python
# change our legend

fig, ax =plt.subplots(figsize=(5,3))
ax.plot(x,y1, label="Person 1")
ax.plot(x,y2, label="Person 2")
ax.legend(loc ='lower right',frameon = True, fontsize = 12,ncol = 2,shadow = True)
```

```python
# change the legend position

fig, ax =plt.subplots(figsize=(5,3))
ax.plot(x,y1, label="Person 1")
ax.plot(x,y2, label="Person 2")
ax.legend(loc ='upper left',
          bbox_to_anchor=(1,1))
```

```python
# Add annotations
fig, ax =plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label="Person 1")
ax.scatter(x,y2, label="Person 2")
ax.legend(loc ='lower right')
ax.text(10,95,"This value is important!",
        ha="center",
        color="red",
        size=15)
```

```python
fig, ax= plt.subplots()
ax.axis([0,10,0,10])

ax.text(1,5,". Data: (1,5)", transform=ax.transData)
ax.text(0.5,0.1,".Axes: (0.5,0.1)", transform=ax.transAxes)
ax.text(0.2,0.2,".Figure:(0.2,0.2)", transform=fig.transFigure)
```

```python
# Use annotate function

fig, ax =plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label="Person 1")
ax.scatter(x,y2, label="Person 2")
ax.legend(loc ='lower right')
ax.annotate("Ths is important!", xy=(10,95), xytext=(20,94), arrowprops=dict(facecolor='black',shrink=0.05))
```

```python
# Change our arrows
fig, ax =plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label="Person 1")
ax.scatter(x,y2, label="Person 2")
ax.legend(loc ='lower right')
ax.annotate("Ths is important!", xy=(10,95), xytext=(20,94), arrowprops=dict(facecolor='red',arrowstyle="wedge"))
```

```python
# Remove ticks and labels
fig, ax =plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label="Person 1")
ax.scatter(x,y2, label="Person 2")
ax.legend(loc ='lower right')

ax.yaxis.set_major_locator(plt.NullLocator()) # Locator to remove ticks and labels
ax.xaxis.set_major_formatter(plt.NullFormatter()) #formatter to remove only labels
```

```python
# Limit tick
fig, ax =plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label="Person 1")
ax.scatter(x,y2, label="Person 2")
ax.legend(loc ='lower right')

ax.xaxis.set_major_locator(plt.MaxNLocator(3)) # limit the axis x to 3 ticks
```

```python
# Tick mark intervals
fig, ax =plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label="Person 1")
ax.scatter(x,y2, label="Person 2")
ax.legend(loc ='lower right')

ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # Interval
```

```python
#modify the axis labels

fig, ax =plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label="Person 1")
ax.scatter(x,y2, label="Person 2")
ax.legend(loc ='lower right')

plt.xticks(rotation=45, ha='right', fontsize=20, font= "serif" , color='red' )
```

```python
#modify the axis labels
font1={'family':'serif','color':'indigo'}
fig, ax =plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label="Person 1")
ax.scatter(x,y2, label="Person 2")
ax.legend(loc ='lower right')

plt.xticks(rotation=45, ha='right', fontsize=20, font= "serif" , color='red' )
plt.xlabel('Shiny New X Axis!', fontsize=18, fontdict=font1)
```
