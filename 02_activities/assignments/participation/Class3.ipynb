```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x1=np.arange(50)
y1=np.random.randint(0,75,50)
x2=np.array(["luffy","Zoro","Nami","Usopp","Sanji"])
y2=np.array([110,180,240,99,220])
```

```python
plt.style.use('ggplot')
fig, (ax1,ax2) =plt.subplots(figsize=(7,3),nrows=1,ncols=2)
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
ax1.annotate("This is important!", xy=(10,55), xytext=(20,34), arrowprops=dict(facecolor='black',shrink=0.05))
ax1.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.xticks(rotation=45, ha='right', fontsize=20, font= "serif" , color='blue' )
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                   ['ax2','ax3']],figsize=(7,4))

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                   ['ax2','ax3']],figsize=(7,4), layout="constrained")

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel("A big label", fontsize=10,color="Blue")
someaxes["ax2"].set_xlabel("Another label", fontsize=10,color="Blue")
someaxes["ax3"].set_xlabel("label 2 : 2 Fast 2 Furious", fontsize=10,color="Blue")
```
