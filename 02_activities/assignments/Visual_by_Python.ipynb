```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("02_activities/assignments/Unemployement_rate.csv")
df.head()
```

```python

# Filter data for Ontario, specified years, and occupations
filtered_df = df[
    (df['Timeseries'] >= 2010) & 
    (df['Prov'] == 'Ontario') & 
    (df['OCCUPATION'].isin([
        ' Management', 
        'NOC skill level B: occupations usually require college education or apprenticeship training', 
        'NOC skill level C: occupations usually require secondary school and/or occupation-specific training',
        'NOC skill level D: on-the-job training is usually provided for these occupations']))
]

sns.barplot(
    data=filtered_df,
    x='Unemployment rate',
    y='Timeseries',
    hue='OCCUPATION',
    orient='h',  # Horizontal bars
    palette='Set2'
)
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1))
plt.title('Unemployment Rate (2010-2019) by Occupation in Ontario', fontsize=16)
plt.show()
```
